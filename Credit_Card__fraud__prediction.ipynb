**Here from Dataset seen that there is unbalanced data between two variables**

```python
#Establish 2 variables i.e Fraud and Non_fraud represented by 1 and 0 respectively

non_fraud=cc[cc.Class==0]
fraud=cc[cc.Class==1]
```

```python
a=len(cc[cc.Class==0])
b=len(cc[cc.Class==1])
```

Under-Sampling
Build a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions
Number of Fraudulent Transactions -> 108


```python
non_fraud_sample=non_fraud.sample(n=108)
```

Concatenating two Dataframes

```python
new_dataset=pd.concat([non_fraud_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()   # here we changed the unbalanced data into balanced
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into features and target

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
print(X.shape)
print(Y.shape)
```

Split the data into training and testing Data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

**Model Training : Logistic Regression**

```python
model =LogisticRegression()
```

```python
#training the model with training data
model.fit (X_train,Y_train)
```

```python
cal_score=model.score(X_test,Y_test)*100
cal_score
```
